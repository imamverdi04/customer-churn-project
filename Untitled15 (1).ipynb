```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.utils import class_weight
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/müşteritkisi (Churn dataset).csv')
data.head()
```

```python
data = data.iloc[:, :-1]
grouped_columns = {
    'int': data.select_dtypes(include='int').columns.tolist(),
    'float': data.select_dtypes(include='float').columns.tolist(),
    'object': data.select_dtypes(include='object').columns.tolist(),
    'bool': data.select_dtypes(include='bool').columns.tolist()
}

grouped_columns
```

```python
# Yaşlı təbəqəni object tipinə çevirirəm ki, kateqoriya kimi işləyə bilim
data['Yaşlı təbəqə'] = data['Yaşlı təbəqə'].astype(object)
```

```python
print("Datasetdə boş qalan yerlər:")
print(data.isnull().sum())
```

```python
data.head()
```

```python
# Kateqoriyalı sütunlardakı boş yerləri doldururam
# Bunun üçün object tipində olan sütunları seçirəm
categorical_columns = data.select_dtypes(include=['object']).columns
for col in categorical_columns:
    # Gender sütunu üçün ən çox rast gəlinən dəyəri istifadə edirəm cünki cinsiyyət təsadüfi ola bilməz
    if col == 'Gender':
        mode_value = data[col].mode()[0]
        data[col] = data[col].fillna(mode_value)
    else:
        # Digər kateqoriyalar üçün 'Unknown' yazıram ki,
        # məlumatın çatışmadığını bilim
        data[col] = data[col].fillna('Unknown')

# Rəqəmsal sütunlardakı boş yerləri median ilə doldururam
# Orta qiyməti götürürəm ki, kənar qiymətlərin təsiri az olsun
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_columns:
    data[col] = data[col].fillna(data[col].median())

print("\nBoş yerləri doldurduqdan sonrakı vəziyyət:")
print(data.isnull().sum())
```

```python
# Kateqoriyalı dəyişənləri rəqəmsallaşdırıram
# ML alqoritmi mətn işləyə bilmir
for col in categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))
```

```python
# Rəqəmsal xüsusiyyətləri standartlaşdırıram
# Bütün dəyişənləri eyni şkalaya gətirirəm ki, model daha yaxşı işləsin
scaler = StandardScaler()
data[numeric_columns] = scaler.fit_transform(data[numeric_columns])
```
